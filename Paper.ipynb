```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from nltk.tokenize import sent_tokenize
import nltk

# Download required data
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

# Read the text file
with open('abstract.txt', 'r') as f:
    text = f.read()

# Split into sentences (chunks)
sentences = sent_tokenize(text)
print(f"Found {len(sentences)} sentences")

# Vectorize text using TF-IDF
vectorizer = TfidfVectorizer(max_features=100, stop_words='english')
X = vectorizer.fit_transform(sentences)

# Apply PCA
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X.toarray())

# Cluster using K-means
kmeans = KMeans(n_clusters=10, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Plot all PCA component combinations
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# All possible pairs of 4 components
component_pairs = [(0,1), (0,2), (0,3), (1,2), (1,3), (2,3)]

for idx, (i, j) in enumerate(component_pairs):
    scatter = axes[idx].scatter(X_pca[:, i], X_pca[:, j], c=clusters, cmap='tab10', alpha=0.7)
    axes[idx].set_title(f'PC{i+1} vs PC{j+1}')
    axes[idx].set_xlabel(f'PC{i+1}')
    axes[idx].set_ylabel(f'PC{j+1}')
    axes[idx].grid(True, alpha=0.3)

# Add colorbar to the last subplot
plt.colorbar(scatter, ax=axes[-1])
plt.tight_layout()
plt.show()

# Print cluster info
for i in range(10):
    cluster_sentences = [sentences[j] for j in range(len(sentences)) if clusters[j] == i]
    print(f"\nCluster {i} ({len(cluster_sentences)} sentences):")
    print(cluster_sentences[0][:200] + "..." if cluster_sentences else "Empty cluster")
```

```python
from sklearn.decomposition import LatentDirichletAllocation


lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda_topics = lda.fit_transform(X)

# Print top words per topic
feature_names = vectorizer.get_feature_names_out()
for i, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[-10:][::-1]]
    print(f"Topic {i}: {', '.join(top_words)}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns

sim_matrix = cosine_similarity(X.toarray())
plt.figure(figsize=(12,5))
sns.heatmap(sim_matrix, cmap='viridis')
plt.title("Sentence Similarity Heatmap")
plt.xlabel("Sentence Index")
plt.ylabel("Sentence Index")
plt.show()
```
